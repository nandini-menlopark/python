```python
import pandas

class school (object):
    def __init__(self):
        self.id = 0
        self.data = {'Name': {}, 'Age': {}, 'Grade': {}}
    
    def addStudent(self, name, age, grade):
        self.id = self.id + 1
        self.data['Name'] [self.id] = name
        self.data['Age'] [self.id] = age
        self.data['Grade'] [self.id] = grade
        
    def getData(self):
        return (pandas.DataFrame(self.data))
    
    def removeStudent(self, id): 
        self.data = pandas.DataFrame(self.data).drop([id]).to_dict()
    

menlopark = school()

menlopark.addStudent('Nandini', 10, '5th')
menlopark.addStudent('Sami', 10, '5th')
menlopark.addStudent('Mayur', 5, 'KG')
menlopark.addStudent('Timtim', 5, 'KG')
menlopark.addStudent('Daksh', 7, '3rd')
menlopark.addStudent('Saranya', 10, '5th')
menlopark.addStudent('Heerva', 8, '3rd')
menlopark.addStudent('Aarna', 9, '5th')

df = menlopark.getData()
```
